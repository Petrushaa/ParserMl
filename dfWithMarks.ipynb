```python
import pandas as pd
import numpy as np

df = pd.read_csv("finam_news.csv")

df
```

```python
df["price_7d"] = pd.to_numeric(df["price_7d"], errors="coerce")
df["price_news"] = pd.to_numeric(df["price_news"], errors="coerce")
df["price_1d"] = pd.to_numeric(df["price_7d"], errors="coerce")
```

```python
df["change"] = (df["price_7d"] - df["price_news"]) / df["price_news"] * 100
df["change"]
```

```python
df["mark"] = np.select(
    [df["change"] >= 1, df["change"] <= -1],
    [1, -1],
    default=0
)

df
```

```python
cols_to_drop = ["price_news", "price_1d", "price_7d"]
df = df.drop(columns=cols_to_drop)
df
```

```python
df.to_csv("news_MarksChangesOnly.csv", index=False, encoding="utf-8")
```
